```python
from sklearn.feature_selection import mutual_info_classif

# Assuming X_train is a pandas DataFrame
discrete_features_indices = [X_train.columns.get_loc(col) for col in discrete_features if col in X_train.columns]
```

```python
def make_mi_scores(X, y, discrete_features):
  result = mutual_info_classif(X_train, y_train, discrete_features=discrete_features_indices)
  result = pd.Series(result, name="MI Scores", index=X_train.columns)
  result = result.sort_values(ascending=False)
  return result
```

```python
miscores = make_mi_scores(X_train, y_train, discrete_features)
```

```python
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")


plt.figure(dpi=100, figsize=(8, 5))
plot_mi_scores(miscores)
```

```python
selected_features = miscores[miscores > 0.15].index
```

```python
selected_features
```

```python
# prompt: implement the best suited classification algorithm to attain the maximum accuracy in predicting the Target feature using the selected_features

import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Train a Logistic Regression model
logreg = LogisticRegression()
logreg.fit(X_train[selected_features], y_train)

# Predict on the validation set
y_pred_logreg = logreg.predict(X_valid[selected_features])

# Calculate the accuracy
accuracy_logreg = accuracy_score(y_valid, y_pred_logreg)

# Train a Random Forest model
rf = RandomForestClassifier()
rf.fit(X_train[selected_features], y_train)

# Predict on the validation set
y_pred_rf = rf.predict(X_valid[selected_features])

# Calculate the accuracy
accuracy_rf = accuracy_score(y_valid, y_pred_rf)

# Print the accuracies
print(f"Logistic Regression Accuracy: {accuracy_logreg:.4f}")
print(f"Random Forest Accuracy: {accuracy_rf:.4f}")

# Select the model with the best accuracy
if accuracy_logreg > accuracy_rf:
    best_model = logreg
else:
    best_model = rf

# Use the best model to predict on the test set
y_pred_test = best_model.predict(dftest[selected_features])

# # Prepare the submission file
# submission_df = pd.DataFrame({'id': dftest['id'], 'Target': y_pred_test})
# submission_df.to_csv('submission.csv', index=False)
```

```python
# prompt: implement all the classification algorithms to attain the maximum accuracy in predicting the Target feature using the selected_features

import pandas as pd
from sklearn.ensemble import VotingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier

# Define the individual classifiers
estimators = list()

estimators.append(('Logistic', LogisticRegression()))
estimators.append(('KNN', KNeighborsClassifier()))
estimators.append(('CART', DecisionTreeClassifier()))
estimators.append(('GNB', GaussianNB()))
estimators.append(('SVC', SVC()))
estimators.append(('XGB', XGBClassifier()))

# Create the ensemble model
ensemble = VotingClassifier(estimators)

# Train the ensemble model
ensemble.fit(X_train[selected_features], y_train)

# Predict on the validation set
y_pred_ensemble = ensemble.predict(X_valid[selected_features])

# Calculate the accuracy
accuracy_ensemble = accuracy_score(y_valid, y_pred_ensemble)

# Print the accuracy
print(f"Ensemble Accuracy: {accuracy_ensemble:.4f}")

# Select the best model based on the highest accuracy
best_model = ensemble if accuracy_ensemble > accuracy_logreg and accuracy_ensemble > accuracy_rf else (logreg if accuracy_logreg > accuracy_rf else rf)

# # Use the best model to predict on the test set
# y_pred_test = best_model.predict(dftest[selected_features])

# # Prepare the submission file
# submission_df = pd.DataFrame({'id': dftest['id'], 'Target': y_pred_test})
# submission_df.to_csv('submission.csv', index=False)
```

```python
# Use the best model to predict on the test set
y_pred_test = best_model.predict(dftest[selected_features])

# Prepare the submission file
submission_df = pd.DataFrame({'id': dftest['id'], 'Target': y_pred_test})
submission_df.to_csv('submission.csv', index=False)
```
